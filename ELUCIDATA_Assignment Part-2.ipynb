# Import Required Libraries

```python
import requests
 and I love youfrom bs4 import BeautifulSoup
```

# Defining gene id and attributes to be extracted

```python
arr=["GSE63312","GSE78224,","GSE74018", "GSE50734", "GSE114644", "GSE60477", "GSE53599",
"GSE80582", "GSE109493", "GSE35200"]
a=["Status","Title","Distribution","Organism","Experiment type","Summary","Overall design","Contributor(s)","Citation(s)"]
data=[]
```

# Function to annotate the summary attribute by BERN API

```python

def query_raw(text, url="https://bern.korea.ac.kr/plain"):
    return requests.post(url, data={'sample_text': text}).json()

```

# Writing the scraping code and iterating for all gene id's

```python
for i in arr:
    res=requests.get('https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc='+str(i))
    soup=BeautifulSoup(res.text,"html.parser")
    d={}
    for j in a:
        body=soup.find(text=j)
        if(body!=None):
            d[str(j)]=body.findNext('td').text
    data.append(d)
```

# Appending annotations to the data extracted

```python
for i in data:
    if "Summary" in i.keys():
        i['annotate']=query_raw(i["Summary"])
        
```

# Printing the Data

```python
for i in data:
    print(i)
```

# Creating and appending it to the Database

```python
from pymongo import MongoClient
  
try:
    conn = MongoClient()
    print("Connected successfully!!!")
except:  
    print("Could not connect to MongoDB")
  
# database
db = conn.database
  
# Created or Switched to collection names: my_gfg_collection
collection = db.databsae
for i in data:
    rec_id=collection.insert_one(i)
    print("Data inserted with record ids",rec_id)
```
